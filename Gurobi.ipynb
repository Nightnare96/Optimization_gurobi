```python
 #pip install gurobipy
import gurobipy as gp
from gurobipy import GRB


def print_solution(model):
    model.optimize()
    if model.status == GRB.OPTIMAL:
        print("\nOptimal Objective =", model.ObjVal)
        for v in model.getVars():
            if abs(v.X) > 1e-6:
                print(f"{v.VarName} = {v.X}")
    else:
        print("No optimal solution found. Status:", model.status)
```

```python
# Questions-1 (Two Variable Maximization )
"""
Maximize: Z = 5x + 4y
6x + 4y <= 24
x + 2y <= 6
x, y >= 0

"""
def Two_Variable_Maximization():
    m = gp.Model("Q1_LP")
    x = m.addVar(lb=0, name="x")
    y = m.addVar(lb=0, name="y")

    m.setObjective(5*x + 4*y, GRB.MAXIMIZE)
    m.addConstr(6*x + 4*y <= 24)
    m.addConstr(x + 2*y <= 6)

    print_solution(m)
Two_Variable_Maximization()
```

```python
# Question-2 (LP — Resource Allocation (3 Variables))
"""
Maximize: Z = 3x1 + 5x2 + 4x3
2x1 + 3x2 + x3 <= 14
4x1 + x2 + 2x3 <= 14
3x1 + 4x2 + 2x3 <= 14
x1, x2, x3 >= 0
"""

def Resource_alocation():
    m = gp.Model("Q2_LP")
    x = m.addVars(3, lb=0, name="x")

    m.setObjective(3*x[0] + 5*x[1] + 4*x[2], GRB.MAXIMIZE)
    m.addConstr(2*x[0] + 3*x[1] + x[2] <= 14)
    m.addConstr(4*x[0] + x[1] + 2*x[2] <= 14)
    m.addConstr(3*x[0] + 4*x[1] + 2*x[2] <= 14)

    print_solution(m)
Resource_alocation()
```

```python
# Question-3 (Cost Minimization)
"""
Minimize: C = 2x + 3y + z
x + y + z >= 10
2x + y >= 8
x + 3z >= 6
x, y, z >= 0
"""
def Cost_Minimization():
    m = gp.Model("Q3_LP")
    x = m.addVar(lb=0, name="x")
    y = m.addVar(lb=0, name="y")
    z = m.addVar(lb=0, name="z")

    m.setObjective(2*x + 3*y + z, GRB.MINIMIZE)
    m.addConstr(x + y + z >= 10)
    m.addConstr(2*x + y >= 8)
    m.addConstr(x + 3*z >= 6)

    print_solution(m)
Cost_Minimization()
```

```python
# Question-4 (Knapsack)
"""
Maximize: Z = 10x1 + 6x2 + 12x3 + 7x4 + 9x5
5x1 + 3x2 + 7x3 + 4x4 + 6x5 <= 14 (weight limit)
x1, x2, x3, x4, x5 in {0, 1}
"""
def Knapsack():
    m = gp.Model("Q4_LP")
    x = m.addVars(5, vtype=GRB.BINARY, name="x")

    values = [10, 6, 12, 7, 9]
    weights = [5, 3, 7, 4, 6]

    m.setObjective(gp.quicksum(values[i]*x[i] for i in range(5)), GRB.MAXIMIZE)
    m.addConstr(gp.quicksum(weights[i]*x[i] for i in range(5)) <= 14)

    print_solution(m)

Knapsack()
```

```python
# Question-5 (Quadratic Objective)
"""
Minimize: f = x^2 + 2y^2 - 4x - 6y
x + y <= 4
x + 3y <= 6
x, y >= 0
"""

def Quadratic_obj():
    m = gp.Model("Q5_LP")
    x = m.addVar(lb=0, name="x")
    y = m.addVar(lb=0, name="y")

    m.setObjective(x*x + 2*y*y - 4*x - 6*y, GRB.MINIMIZE)
    m.addConstr(x + y <= 4)
    m.addConstr(x + 3*y <= 6)

    print_solution(m)
Quadratic_obj()
```

```python
# Question-6 ( Assignment Problem)
"""
Minimize: Z = 9x11+2x12+7x13 + 3x21+6x22+4x23 + 1x31+8x32+5x33
x11+x12+x13 = 1, x21+x22+x23 = 1, x31+x32+x33 = 1 (each worker assigned once)
x11+x21+x31 = 1, x12+x22+x32 = 1, x13+x23+x33 = 1 (each job covered once)
"""
def Assignment_Problem():
    m = gp.Model("Q6_LP")
    cost = [
        [9, 2, 7],
        [3, 6, 4],
        [1, 8, 5]
    ]
    x = m.addVars(3, 3, vtype=GRB.BINARY, name="x")

    m.setObjective(gp.quicksum(cost[i][j]*x[i, j] for i in range(3) for j in range(3)), GRB.MINIMIZE)

    for i in range(3):
        m.addConstr(gp.quicksum(x[i, j] for j in range(3)) == 1)

    for j in range(3):
        m.addConstr(gp.quicksum(x[i, j] for i in range(3)) == 1)

    print_solution(m)
Assignment_Problem()
```

```python
# Question-7 (Transportation Problem)
"""
Minimize: Z = 2x11+3x12+1x13 + 5x21+4x22+8x23 + 5x31+6x32+8x33
x11+x12+x13 = 30 (supply S1)
x21+x22+x23 = 40 (supply S2)
x31+x32+x33 = 50 (supply S3)
x11+x21+x31 = 25 (demand D1)
x12+x22+x32 = 35 (demand D2)
x13+x23+x33 = 60 (demand D3)
all xij >= 0
"""

def Transportation_Problem():
    m = gp.Model("Q7_LP")
    cost = [
        [2, 3, 1],
        [5, 4, 8],
        [5, 6, 8]
    ]

    supply = [30, 40, 50]
    demand = [25, 35, 60]

    x = m.addVars(3, 3, lb=0, name="x")

    m.setObjective(gp.quicksum(cost[i][j]*x[i, j] for i in range(3) for j in range(3)), GRB.MINIMIZE)

    for i in range(3):
        m.addConstr(gp.quicksum(x[i, j] for j in range(3)) == supply[i])

    for j in range(3):
        m.addConstr(gp.quicksum(x[i, j] for i in range(3)) == demand[j])

    print_solution(m)
Transportation_Problem()
```

```python
# Question-8 (Nonlinear Constraints)
"""
Minimize: f = (x-2)^2 + (y-3)^2
x^2 + y^2 <= 25
x + y >= 4
x, y >= 0
"""

def Nonlinear_Constraints():
    m = gp.Model("Q8_LP")
    x = m.addVar(lb=0, name="x")
    y = m.addVar(lb=0, name="y")
    m.setObjective((x - 2)*(x - 2) + (y - 3)*(y - 3), GRB.MINIMIZE)
    m.addConstr(x*x + y*y <= 25)
    m.addConstr(x + y >= 4)

    print_solution(m)

Nonlinear_Constraints()
```

```python
# Question-9 ( Facility Selection )
"""
Minimize: Z = 10y1+15y2+12y3 + 3x11+7x12+5x13 + 6x21+4x22+8x23
x11+x12+x13 >= 20 (demand client 1)
x21+x22+x23 >= 30 (demand client 2)
x11+x21 <= 40*y1, x12+x22 <= 35*y2, x13+x23 <= 30*y3
y1, y2, y3 in {0,1}; all xij >= 0
"""

def Facility_Selection():
    m = gp.Model("Q9_LP")
    y = m.addVars(3, vtype=GRB.BINARY, name="y")
    x = m.addVars(2, 3, lb=0, name="x")

    fixed_cost = [10, 15, 12]
    trans_cost = [
        [3, 7, 5],
        [6, 4, 8]
    ]
    capacity = [40, 35, 30]

    m.setObjective(
        gp.quicksum(fixed_cost[j]*y[j] for j in range(3)) +
        gp.quicksum(trans_cost[i][j]*x[i, j] for i in range(2) for j in range(3)),
        GRB.MINIMIZE
    )

    m.addConstr(gp.quicksum(x[0, j] for j in range(3)) >= 20)
    m.addConstr(gp.quicksum(x[1, j] for j in range(3)) >= 30)

    for j in range(3):
        m.addConstr(x[0, j] + x[1, j] <= capacity[j] * y[j])

    print_solution(m)

Facility_Selection()
```

```python
# Question-10 (Diet Problem)
"""
Minimize: Cost = 0.60*x1 + 0.35*x2 + 0.80*x3 + 0.50*x4 ($/serving)
3x1 + 2x2 + 4x3 + x4 >= 10 (protein, g)
2x1 + 4x2 + x3 + 3x4 >= 12 (carbs, g)
x1 + x2 + x3 + x4 >= 4 (min servings)
x1, x2, x3, x4 >= 0
"""

def Diet_Problem():
    m = gp.Model("Q10_LP")
    x = m.addVars(4, lb=0, name="x")
    cost = [0.60, 0.35, 0.80, 0.50]

    m.setObjective(gp.quicksum(cost[i]*x[i] for i in range(4)), GRB.MINIMIZE)

    m.addConstr(3*x[0] + 2*x[1] + 4*x[2] + x[3] >= 10)
    m.addConstr(2*x[0] + 4*x[1] + x[2] + 3*x[3] >= 12)
    m.addConstr(gp.quicksum(x[i] for i in range(4)) >= 4)

    print_solution(m)
Diet_Problem()
```

```python
# Question-11 (Capital Budgeting (MIP ))
"""
Maximize: NPV = 14x1 + 17x2 + 17x3 + 15x4 + 40x5 ($ millions)
12x1 + 54x2 + 6x3 + 6x4 + 30x5 <= 48 (year 1 budget)
3x1 + 7x2 + 6x3 + 2x4 + 35x5 <= 20 (year 2 budget)
x2 + x4 <= 1 (mutually exclusive projects)
x1, x2, x3, x4, x5 in {0, 1}
"""

def Cap_Budget ():
    m = gp.Model("Q11_Lp")

    x = m.addVars(5, vtype=GRB.BINARY, name="x")

    npv = [14, 17, 17, 15, 40]
    budget1 = [12, 54, 6, 6, 30]
    budget2 = [3, 7, 6, 2, 35]

    m.setObjective(gp.quicksum(npv[i]*x[i] for i in range(5)), GRB.MAXIMIZE)

    m.addConstr(gp.quicksum(budget1[i]*x[i] for i in range(5)) <= 48)
    m.addConstr(gp.quicksum(budget2[i]*x[i] for i in range(5)) <= 20)
    m.addConstr(x[1] + x[3] <= 1)

    print_solution(m)
Cap_Budget()
```

```python
# Question-12 (Integer LP (Production Lots))
"""
Maximize: Profit = 25x + 30y
20x + 30y <= 1200 (machine hours)
2x + y <= 80 (labor hours)
x <= 40 (demand cap x)
y <= 30 (demand cap y)
x, y in Z+ (non-negative integers)
"""

def Production_Lots():
    m = gp.Model("Q12_Lp")
    x = m.addVar(vtype=GRB.INTEGER, lb=0, name="x")
    y = m.addVar(vtype=GRB.INTEGER, lb=0, name="y")

    m.setObjective(25*x + 30*y, GRB.MAXIMIZE)

    m.addConstr(20*x + 30*y <= 1200)
    m.addConstr(2*x + y <= 80)
    m.addConstr(x <= 40)
    m.addConstr(y <= 30)

    print_solution(m)
Production_Lots()
```

```python
# Question-13 (Network Shortest Path)
"""
Minimize: Z = 4x12+2x13+5x14 + 3x23+1x24 + 6x34 + 2x35 + 3x45
x12+x13+x14 = 1 (flow out of source node 1)
-x12+x23+x24 = 0 (flow balance node 2)
-x13-x23+x34+x35 = 0 (flow balance node 3)
-x14-x24-x34+x45 = 0 (flow balance node 4)
-x35-x45 = -1 (flow into sink node 5)
all xij >= 0
"""

def Network_Shortest_Path():
    m = gp.Model("Q13_LP")
    arcs = {
        (1, 2): 4,
        (1, 3): 2,
        (1, 4): 5,
        (2, 3): 3,
        (2, 4): 1,
        (3, 4): 6,
        (3, 5): 2,
        (4, 5): 3
    }

    x = m.addVars(arcs.keys(), lb=0, name="x")

    m.setObjective(gp.quicksum(arcs[a]*x[a] for a in arcs), GRB.MINIMIZE)

    m.addConstr(x[1, 2] + x[1, 3] + x[1, 4] == 1)
    m.addConstr(-x[1, 2] + x[2, 3] + x[2, 4] == 0)
    m.addConstr(-x[1, 3] - x[2, 3] + x[3, 4] + x[3, 5] == 0)
    m.addConstr(-x[1, 4] - x[2, 4] - x[3, 4] + x[4, 5] == 0)
    m.addConstr(-x[3, 5] - x[4, 5] == -1)

    print_solution(m)
Network_Shortest_Path()
```

```python
# Question-14 (Shift Scheduling)
"""
Minimize: Cost = 170x1+160x2+175x3+180x4+195x5
x1+x2 >= 6 (Mon demand)
x2+x3 >= 8 (Tue demand)
x3+x4 >= 9 (Wed demand)
x4+x5 >= 7 (Thu demand)
x1+x5 >= 5 (Fri demand)
x1,x2,x3,x4,x5 >= 0 and integer
"""

def Shift_Scheduling():
    m = gp.Model("Q14_LP")
    x = m.addVars(5, vtype=GRB.INTEGER, lb=0, name="x")
    cost = [170, 160, 175, 180, 195]

    m.setObjective(gp.quicksum(cost[i]*x[i] for i in range(5)), GRB.MINIMIZE)

    m.addConstr(x[0] + x[1] >= 6)
    m.addConstr(x[1] + x[2] >= 8)
    m.addConstr(x[2] + x[3] >= 9)
    m.addConstr(x[3] + x[4] >= 7)
    m.addConstr(x[0] + x[4] >= 5)

    print_solution(m)
Shift_Scheduling()
```

```python
# Question-15 ( Shadow Price )
"""
Maximize: Z = 40x1 + 30x2
x1 + x2 <= 40 (labor hours, RHS = 40)
2x1 + x2 <= 60 (machine hours, RHS = 60)
x1 <= 30 (market limit x1)
x1, x2 >= 0
"""
def Shadow_Price():
    m = gp.Model("Q15_LP")
    x1 = m.addVar(lb=0, name="x1")
    x2 = m.addVar(lb=0, name="x2")

    m.setObjective(40*x1 + 30*x2, GRB.MAXIMIZE)

    c1 = m.addConstr(x1 + x2 <= 40, name="labor")
    c2 = m.addConstr(2*x1 + x2 <= 60, name="machine")
    c3 = m.addConstr(x1 <= 30, name="market")

    m.optimize()

    if m.status == GRB.OPTIMAL:
        print("\nOptimal Objective =", m.ObjVal)
        print("x1 =", x1.X)
        print("x2 =", x2.X)

        print("\nShadow Prices:")
        for c in [c1, c2, c3]:
            print(c.ConstrName, "Pi =", c.Pi)
Shadow_Price()
```

```python
# Question-16 (Lot Sizing (3 Periods))
"""
Minimize: Cost = 100y1+100y2+100y3 + 5I1+5I2+5I3 + 2x1+2x2+2x3
x1 - I1 = 40 (demand period 1: d1=40)
x2 + I1 - I2 = 30 (demand period 2: d2=30)
x3 + I2 - I3 = 50 (demand period 3: d3=50)
x1 <= 120*y1, x2 <= 120*y2, x3 <= 120*y3 (setup links)
x1,x2,x3,I1,I2,I3 >= 0; y1,y2,y3 in {0,1}
"""

def Lot_Sizing():
    m = gp.Model("Q16_LP")
    x = m.addVars(3, lb=0, name="x")
    I = m.addVars(3, lb=0, name="I")
    y = m.addVars(3, vtype=GRB.BINARY, name="y")

    m.setObjective(
        gp.quicksum(100*y[t] + 5*I[t] + 2*x[t] for t in range(3)),
        GRB.MINIMIZE
    )

    m.addConstr(x[0] - I[0] == 40)
    m.addConstr(x[1] + I[0] - I[1] == 30)
    m.addConstr(x[2] + I[1] - I[2] == 50)

    for t in range(3):
        m.addConstr(x[t] <= 120*y[t])

    print_solution(m)
Lot_Sizing()
```

```python
# Question-17 (Game Theory (Zero-Sum Game))
"""
Maximize: v (value of the game)
3p1 + 1p2 + 0p3 >= v (column player strategy 1)
0p1 + 2p2 + 4p3 >= v (column player strategy 2)
2p1 + 3p2 + 1p3 >= v (column player strategy 3)
p1 + p2 + p3 = 1
p1, p2, p3 >= 0
"""

def Game_Theory():
    m = gp.Model("Q17_LP")
    p = m.addVars(3, lb=0, name="p")
    v = m.addVar(lb=-GRB.INFINITY, name="v")

    m.setObjective(v, GRB.MAXIMIZE)

    m.addConstr(3*p[0] + 1*p[1] + 0*p[2] >= v)
    m.addConstr(0*p[0] + 2*p[1] + 4*p[2] >= v)
    m.addConstr(2*p[0] + 3*p[1] + 1*p[2] >= v)

    m.addConstr(p[0] + p[1] + p[2] == 1)

    print_solution(m)
Game_Theory()
```
